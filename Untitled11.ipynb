```python
import pandas as pd
dec=pd.read_csv('SOCR-HeightWeight (1).csv')
print(dec)
```

```python


import pandas as pd
def1=pd.read_csv('class1.csv')
def2=pd.read_csv('class2.csv')
print(def1)
print(def2)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
fig1=px.scatter(def1,x="Student Number",y="Marks",title="class1")
fig2=px.scatter(def2,x="Student Number",y="Marks",title="class1")
marks1=def1["Marks"]
marks2=def2["Marks"]
average1=marks1.mean()
average2=marks2.mean()
print("class1 average",average1)
print("class2 average",average2)
length=marks1.count()
y0=average1
y1=average2
x0=1
x1=length
fig1.update_layout(shapes=[dict(
type="line",y0=y0,y1=y1,x0=x0,x1=x1
)])
fig1.update_yaxes(rangemode="tozero")
fig2.update_yaxes(rangemode="tozero")
fig1.show()
fig2.show()
print("Standard deviation of Marks of class1 from centeral tendancy",marks1.std())
print("Standard deviation of Marks of class from centeral tendancy",marks2.std())
```
